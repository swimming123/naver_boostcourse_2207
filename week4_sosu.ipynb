```python
def eratosthenes(n): # Eratosthenes's seive
  sieve = [True] * (n+1) #declare list of 'n' ea True, True means Prime
  for i in range(2, n+1):
    # if i is prime, i's muliple is not prime. -> i : True, i's multiple : False
    for j in range(i+i, n+1, i):
      sieve[j] = False
    # num of True = num of Prime
  result = []
  for i in range(2, n+1):
    if sieve[i] == True :
      result.append(i)
  return result
def between_prime():
  start = int(input("start number?: "))
  end = int(input("end number?: "))
  return len(eratosthenes(end)) - len(eratosthenes(start))
print(between_prime())
```

```python
n = int(input('첫번째 수 입력: '))
m = int(input('두번째 수 입력: '))

def prime(n):
  sieve = [True] * n #에라토스테네스의 체 초기화: 자연수 n에 True 설정 (자연수 n을 소수로 간주함)
  m = int(n ** 0.5)
  for i in range(2, m+1): #1은 제외하고 2부터 
    if sieve[i] == True : #i가 소수인 경우
      for j in range(i+i, n, i): #i 이후 n까지의 i의 배수인 j들을 false 판정
        sieve[j] = False
  return [i for i in range(2, n) if sieve[i] == True]

print(len(prime(m))-len(prime(n)))
```
