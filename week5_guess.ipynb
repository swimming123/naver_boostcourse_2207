```python
# Q3. 숫자 맞추기 게임을 만들어 볼게요. 컴퓨터가 임의의 숫자를 3개 만들고 우리가 그것을 맞춰보겠습니다.

# 😲조건1 - 숫자는 0 ~ 100까지 숫자를 3개 만듭니다(중복 불가).
# 😲조건2 - 5회, 10회까지 정답을 못맞추면 최솟값, 최대값에 대한 힌트를 줍니다.
# 😲조건3 - 정답을 맞추면, 맞춘 정답이 최솟값인지, 중간값인지, 최댓값인지 알려줍니다.
```

```python
import random

def guess_numbers(numbers) : 
    numbers = numbers
    #횟수 count
    tries = 1
    #사용한 숫자 저장
    used = []
    used.append(999)
    #맞춘 횟수
    dingdongdang = 0
    print('1 ~ 100 사이의 숫자를 예측해 보세요')
    print('종료 : 999')

    while tries > 0 : 
        #시도횟수
        print(' ', end='\n')
        print(tries, '차 시도 ')   
          

        # 숫자 3개 모두 맞추면 종료
        if dingdongdang == 3 : 
            print('숫자를 모두 맞추셨습니다!')
            print('정답은 ', end=" ")
            for j in range(len(numbers)) : 
                print(numbers[j], end=", ") 
            break  
 
        #숫자입력시 게임 시작, 아닐경우 except 출력 후 재입력받음
        try : 
            #숫자 입력
            guess = int(input('숫자를 입력해 주세요 : '))
            # 사용한 숫자인지 확인
            if guess not in used :
                used.append(guess)
                used.sort()
                #print(used)
     
                # 숫자 맞추기, 성공일 때
                if guess == numbers[0] or guess == numbers[1] or guess == numbers[2] : 
                    if guess == numbers[0] : 
                        print('숫자를 맞추셨습니다! {}는 최솟값 입니다! '.format(guess))
                        print('{}번 시도만에 예측 성공'.format(tries))
                        tries += 1
                        dingdongdang += 1
                    elif guess == numbers[1] : 
                        print('숫자를 맞추셨습니다! {}는 중간값 입니다! '.format(guess))
                        print('{}번 시도만에 예측 성공'.format(tries))
                        dingdongdang += 1
                        tries += 1
                    else : 
                        print('숫자를 맞추셨습니다! {}는 최댓값 입니다! '.format(guess))
                        print('{}번 시도만에 예측 성공'.format(tries))
                        dingdongdang += 1
                else : 
                    print(guess, '는 없습니다. 다시 입력해 주세요!')
                    tries += 1

                # 5, 10, ... 번째 마다 힌트
                #최소값
                min_hint = guess > min(numbers) and min(numbers) not in used
                #최대값
                max_hint = guess > max(numbers) and max(numbers) not in used
                #중간값보다 작을 때
                mid_hint = min(numbers) < guess < numbers[1] 

                if tries % 5 == 0 : 
                    if min_hint : 
                        print('hint!')
                        print('***최솟값은 {}보다 작습니다.***'.format(guess))
                    elif max_hint : 
                        print('hint!')
                        print('***최댓값은 {}보다 작습니다***'.format(guess))
                    elif mid_hint : 
                        print('hint!')
                        print('***{}는 최솟값과 중앙값 사이에 있습니다***'.format(guess))
                    else : 
                        print('hint!')
                        print('***{}는 중앙값과 최대값 사이에 있습니다***'.format(guess))
                # 힌트 끝남
                else : 
                    continue
            #종료
            elif guess == 999 : 
              break
            
            else : 
                print('이미 사용한 숫자입니다. 다른 숫자를 입력해 주세요. ')
                continue
            
        #숫자 아닌 값 입력했을 때
        except : 
            print('숫자만 입력해 주세요')
            continue  


#컴퓨터 난수 생성 함수
def com_random() : 
    numbers = random.sample(range(1, 101),3)
    numbers.sort()
    print(numbers)
    guess_numbers(numbers)

com_random()
```
