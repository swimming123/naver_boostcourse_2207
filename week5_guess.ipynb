```python
import random

def guess_numbers(numbers) : 
    numbers = numbers
    #íšŸìˆ˜ count
    tries = 1
    #ì‚¬ìš©í•œ ìˆ«ì ì €ì¥
    used = []
    used.append(999)
    #ë§ì¶˜ íšŸìˆ˜
    dingdongdang = 0
    print('1 ~ 100 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì˜ˆì¸¡í•´ ë³´ì„¸ìš”')
    print('ì¢…ë£Œ : 999')

    while tries > 0 : 
        #ì‹œë„íšŸìˆ˜
        print(' ', end='\n')
        print(tries, 'ì°¨ ì‹œë„ ')   
          

        # ìˆ«ì 3ê°œ ëª¨ë‘ ë§ì¶”ë©´ ì¢…ë£Œ
        if dingdongdang == 3 : 
            print('ìˆ«ìë¥¼ ëª¨ë‘ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤!')
            print('ì •ë‹µì€ ', end=" ")
            for j in range(len(numbers)) : 
                print(numbers[j], end=", ") 
            break  
 
        #ìˆ«ìì…ë ¥ì‹œ ê²Œì„ ì‹œì‘, ì•„ë‹ê²½ìš° except ì¶œë ¥ í›„ ì¬ì…ë ¥ë°›ìŒ
        try : 
            #ìˆ«ì ì…ë ¥
            guess = int(input('ìˆ«ìë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš” : '))
            # ì‚¬ìš©í•œ ìˆ«ìì¸ì§€ í™•ì¸
            if guess not in used :
                used.append(guess)
                used.sort()
                #print(used)
     
                # ìˆ«ì ë§ì¶”ê¸°, ì„±ê³µì¼ ë•Œ
                if guess == numbers[0] or guess == numbers[1] or guess == numbers[2] : 
                    if guess == numbers[0] : 
                        print('ìˆ«ìë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤! {}ëŠ” ìµœì†Ÿê°’ ì…ë‹ˆë‹¤! '.format(guess))
                        print('{}ë²ˆ ì‹œë„ë§Œì— ì˜ˆì¸¡ ì„±ê³µ'.format(tries))
                        tries += 1
                        dingdongdang += 1
                    elif guess == numbers[1] : 
                        print('ìˆ«ìë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤! {}ëŠ” ì¤‘ê°„ê°’ ì…ë‹ˆë‹¤! '.format(guess))
                        print('{}ë²ˆ ì‹œë„ë§Œì— ì˜ˆì¸¡ ì„±ê³µ'.format(tries))
                        dingdongdang += 1
                        tries += 1
                    else : 
                        print('ìˆ«ìë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤! {}ëŠ” ìµœëŒ“ê°’ ì…ë‹ˆë‹¤! '.format(guess))
                        print('{}ë²ˆ ì‹œë„ë§Œì— ì˜ˆì¸¡ ì„±ê³µ'.format(tries))
                        dingdongdang += 1
                else : 
                    print(guess, 'ëŠ” ì—†ìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì…ë ¥í•´ ì£¼ì„¸ìš”!')
                    tries += 1

                # 5, 10, ... ë²ˆì§¸ ë§ˆë‹¤ íŒíŠ¸
                #ìµœì†Œê°’
                min_hint = guess > min(numbers) and min(numbers) not in used
                #ìµœëŒ€ê°’
                max_hint = guess > max(numbers) and max(numbers) not in used
                #ì¤‘ê°„ê°’ë³´ë‹¤ ì‘ì„ ë•Œ
                mid_hint = min(numbers) < guess < numbers[1] 

                if tries % 5 == 0 : 
                    if min_hint : 
                        print('hint!')
                        print('***ìµœì†Ÿê°’ì€ {}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤.***'.format(guess))
                    elif max_hint : 
                        print('hint!')
                        print('***ìµœëŒ“ê°’ì€ {}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤***'.format(guess))
                    elif mid_hint : 
                        print('hint!')
                        print('***{}ëŠ” ìµœì†Ÿê°’ê³¼ ì¤‘ì•™ê°’ ì‚¬ì´ì— ìˆìŠµë‹ˆë‹¤***'.format(guess))
                    else : 
                        print('hint!')
                        print('***{}ëŠ” ì¤‘ì•™ê°’ê³¼ ìµœëŒ€ê°’ ì‚¬ì´ì— ìˆìŠµë‹ˆë‹¤***'.format(guess))
                # íŒíŠ¸ ëë‚¨
                else : 
                    continue
            #ì¢…ë£Œ
            elif guess == 999 : 
              break
            
            else : 
                print('ì´ë¯¸ ì‚¬ìš©í•œ ìˆ«ìì…ë‹ˆë‹¤. ë‹¤ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš”. ')
                continue
            
        #ìˆ«ì ì•„ë‹Œ ê°’ ì…ë ¥í–ˆì„ ë•Œ
        except : 
            print('ìˆ«ìë§Œ ì…ë ¥í•´ ì£¼ì„¸ìš”')
            continue  


#ì»´í“¨í„° ë‚œìˆ˜ ìƒì„± í•¨ìˆ˜
def com_random() : 
    numbers = random.sample(range(1, 101),3)
    numbers.sort()
    print(numbers)
    guess_numbers(numbers)

com_random()
```
